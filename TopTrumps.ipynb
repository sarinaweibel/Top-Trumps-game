```python
import random 
import time
from termcolor import colored

#define class
class Card:
    def __init__(self, Name, Magic, Cunning, Courage, Wiscom, Temper):
        #card values
        self.Name = Name
        self.Magic = Magic
        self.Cunning = Cunning
        self.Courage = Courage
        self.Wisdom = Wisdom
        self.Temper = Temper

deck = [] #create empty list

#open file
fh = open('harry potter (2).txt','r')

#save data, readlines() creates a list of each line of a file
Data=fh.readlines()

#delete first line
first_line = Data.pop(0)


#read file and create cards
for line in Data:
  row = line.split(',') #split rows (strings) into a list
  Name, Magic, Cunning, Courage, Wisdom, Temper = [i.strip() for i in row]
  card = Card(Name, Magic, Cunning, Courage, Wisdom, Temper)
  deck = deck + [ card ]


#close file
fh.close()

#shuffle deck
random.shuffle(deck)


#initialize player and computer deck
playerDeck = []
computerDeck = []

#Deal the cards to the decks
while len(deck) > 0:
  playerDeck.append(deck.pop(0))
  computerDeck.append(deck.pop(0))

#The player will beginn
playerTurn = True
allowedResponses = ["a", "b", "c", "d", "e"]

#main programm
while len(playerDeck) > 0 and len(computerDeck) > 0: #loop as long both player and computer have cards
 
  time.sleep(2) #delay the programm so it's more comfortable to play 
  
  #remove one card from each deck and add it to the gameroundvalue
  playerCard = playerDeck.pop(0) 
  computerCard = computerDeck.pop(0)
  
  # print player card value
  print("\033[1m" "YOUR CARD:" "\033[0m") #printed in bold letters for improved visibility
  print("Name:", playerCard.Name) 
  print("a. Magic:", playerCard.Magic)
  print("b. Cunning:", playerCard.Cunning)
  print("c. Courage:", playerCard.Courage)
  print("d. Wisdom:", playerCard.Wisdom)
  print("e. Temper:", playerCard.Temper)
  
  time.sleep(2)


  #check if its the turn of the player 
  if playerTurn == True:
    #get the player input 
    answer = input("Which attribute do you choose? (a, b, c, d or e): ")
    print('') #space for improved visibility
    #check if inputist valid
    while allowedResponses.count(answer) == 0:
      answer = input("That isn't a valid answer, please try again: ")
    #if it isn't the turn of the player choose random value from the computer card
  else:
    answer = random.choice(allowedResponses)
    print("Computer chooses", answer)
    print('') #space for improved visibility
  
  time.sleep(2)

  print("\033[1m" "COMPUTER CARD" "\033[0m")  #printed in bold letters
  print("Name:", computerCard.Name)
    
  playerWins = True

  #check input and print computercard value
  if answer == "a":
    print("Magic:", computerCard.Magic)
    print('')
    isDraw = (int(playerCard.Magic) == int(computerCard.Magic))         #check if its a draw
    playerWins = (int(playerCard.Magic) > int(computerCard.Magic))      #check if player wins

  elif answer == "b":
    print("Cunning:", computerCard.Cunning)
    print('')
    isDraw = (int(playerCard.Cunning) == int(computerCard.Cunning))     #check if its a draw
    playerWins = (int(playerCard.Cunning) > int(computerCard.Cunning))  #check if player wins
  elif answer == "c":
    print("Courage:", computerCard.Courage)
    print('')
    isDraw = (int(playerCard.Courage) == int(computerCard.Courage))     #check if its a draw
    playerWins = (int(playerCard.Courage) > int(computerCard.Courage))  #check if player wins
  elif answer == "d":
    print("Wisdom:", computerCard.Wisdom)
    print('')
    isDraw = (int(playerCard.Wisdom) == int(computerCard.Wisdom))       #check if its a draw
    playerWins = (int(playerCard.Wisdom) > int(computerCard.Wisdom))    #check if player wins
  elif answer == "e":
    print("Temper:", computerCard.Temper)
    print('')
    isDraw = (int(playerCard.Temper) == int(computerCard.Temper))       #check if its a draw
    playerWins = (int(playerCard.Temper) < int(computerCard.Temper))    #check if player wins

    time.sleep(2) #delay the programm so it's more comfortable to play 

    
  if isDraw:
    print("It's a tie!")
    print('')
    time.sleep(2)
    #add both cards back to the decks
    playerDeck.append(playerCard)
    computerDeck.append(computerCard)
    
  elif playerWins:
    print("\033[92m" "YOU WIN THIS ROUND!" "\033[0m" )  #printed in green letters 
    print('')
    time.sleep(2)
    #add both card to the player deck
    playerDeck.append(playerCard)
    playerDeck.append(computerCard)
    playerTurn = True
  else:
    print("YOU LOSE THIS ROUND!") #printed in red letters 
    print('')
    time.sleep(2)
    #add both cards to the computer deck
    computerDeck.append(computerCard)
    computerDeck.append(playerCard)
    playerTurn = False
    
  time.sleep(2)

#check who won
if len(playerDeck) == 0:
    print("UNFORTUNATELY YOU LOST THE GAME. MAYBE YOU WIN THE NEXT ONE!")
else:
    print("CONGRATULATION YOU WON THE GAME!")
```
